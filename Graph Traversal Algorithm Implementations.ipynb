Dijkstra's Algorithm

```python
#Note: Graph is assumed to be acyclic; no checks are made to ensure acyclic nature of graph. Also, weights are assumed to be nonnegative.

def dijkstra(nodes: List[int], edges: List[List[int]], weights: List[float], source: int, destination: int) -> List[int]:
  numNodes = len(nodes)
  if (numNodes == 0):
    return []
  
  currentNode = source
  paths = [None] * numNodes
  paths[currentNode] = [currentNode]
  minimumTotalCosts = [math.inf] * numNodes
  minimumTotalCosts[currentNode] = 0
  isVisited = [False] * numNodes
  isVisited[currentNode] = True

  def findNextNode(nodes):
      min = math.inf
      nextNode = None
      for i in range(len(nodes)):
        if (minimumTotalCosts[i] < min and isVisited[i] == False):
          nextNode = nodes[i]
          min = minimumTotalCosts[i]
      
      return nextNode
  
  def updateBestPath(currentNode, candidateDestination, weightOfCandidateEdge):
    if (minimumTotalCosts[currentNode] + weightOfCandidateEdge < minimumTotalCosts[candidateDestination]):
        minimumTotalCosts[candidateDestination] = minimumTotalCosts[currentNode] + weightOfCandidateEdge
        paths[candidateDestination] = paths[currentNode] + [candidateDestination]

  while (currentNode != destination):
    candidateEdgesAndIndices = list(filter(lambda element: element[1][0] == currentNode, enumerate(edges)))
    indicesOfCandidateEdges = [x for x,_ in candidateEdgesAndIndices]
    candidateEdges = [y for _, y in candidateEdgesAndIndices]

    for i in range(0, len(candidateEdges)):
      candidateEdge = candidateEdges[i]
      indexOfCandidateEdge = indicesOfCandidateEdges[i]
      candidateDestination = candidateEdge[1]
      weightOfCandidateEdge = weights[indexOfCandidateEdge]
      updateBestPath(currentNode, candidateDestination, weightOfCandidateEdge)
    
    currentNode = findNextNode(nodes)
    isVisited[currentNode] = True
  
  return paths[destination]
```

```python
nodes = range(0, 7)
edges = [[0, 1], [0, 2], [1, 3], [2, 3], [3, 4], [3, 5], [4, 5], [4, 6], [5, 6]]
weights = [2, 6, 5, 8, 10, 15, 6, 2, 6]
source = 3
destination = 6

path = dijkstra(nodes, edges, weights, source, destination)

print(path)
```
